```python
import joblib

model = joblib.load("movie_model.pkl")
vectorizer = joblib.load("vectorizer.pkl")

print("Model loaded successfully ‚úÖ")
```

```python
def predict_review(text):
    vec = vectorizer.transform([text])
    prob = model.predict_proba(vec)[0][1]
    return prob
```

```python
predict_review("Amazing emotional story and acting")
```

```python
from IPython.display import display
import ipywidgets as widgets

text = widgets.Textarea(
    placeholder="Type your movie review here...",
    layout=widgets.Layout(width='80%', height='120px')
)

button = widgets.Button(description="üé¨ Predict")
output = widgets.Output()

def on_click(b):
    with output:
        output.clear_output()
        prob = predict_review(text.value)
        print("Positive Confidence:", round(prob*100,2), "%")
        
        if prob > 0.5:
            print("üéØ Positive Impact")
        else:
            print("‚ö†Ô∏è Negative Impact")

button.on_click(on_click)

display(text, button, output)
```
