```python
data = pd.read_csv('IMDB Dataset.csv')
print(data.head())
```

```python
rows = data.shape[0]
print(f'The dataset contains {rows} rows.') 
```

```python
import string
# Remove punctuation from all text columns
data = data.apply(
    lambda col: 
        col.str.translate(str.maketrans('', '', string.punctuation)) 
            if col.dtype == "object" 
            else col
)

print("Punctuation removed successfully!")
data.head()
```

```python
lower_data = data.apply(
    lambda col:
        col.str.lower()
            if col.dtype == "object"
            else col
)
print("Converted to lowercase successfully!")
lower_data.head()
```

```python
remove_html = lower_data.apply(
    lambda col:
        col.str.replace(r'<.*?>', '', regex=True)
            if col.dtype == "object"
            else col
)
print("HTML tags removed successfully!")
remove_html.head()
```

```python
remove_stopwords = remove_html.copy()
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
remove_stopwords['review'] = remove_stopwords['review'].apply(
    lambda text: ' '.join(
        word for word in text.split() if word not in ENGLISH_STOP_WORDS
    )
)
print("Stop words removed successfully!")
remove_stopwords.head()
```

```python
remove_duplicates = remove_stopwords.drop_duplicates()
print(f"Duplicates removed! New dataset contains {remove_duplicates.shape[0]} rows.")
remove_duplicates.head()
```

```python
data["word_count"] = data['review'].apply(lambda x: len(x.split()))
print("Word count added successfully!")
data.head()
```

```python
data["sentiment_lable"] = data['sentiment'].apply(lambda x: 1 if x == 'positive' else 0)
print("Sentiment label added successfully!")
data.head()
```

```python
data["char_length"] = data['review'].apply(lambda x: len(x))
print("Character length added successfully!")
data.head()
```

```python
data["review_id"] = np.arange(len(data))
print("Review ID added successfully!")
data.head()
```

```python
import re
stop_words = set(ENGLISH_STOP_WORDS)

def clean_text(text):
    text = str(text).lower()
    
    # remove html tags
    text = re.sub(r'<.*?>', ' ', text)
    
    # remove punctuation and numbers
    text = re.sub(r'[^a-z\s]', ' ', text)
    
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # remove stopwords
    words = [w for w in text.split() if w not in stop_words]
    
    return " ".join(words)
data["clean_text"] = data["review"].apply(clean_text)
data.head()

```

```python
themes = {
    "acting": ["actor", "acting", "performance"],
    "story": ["story", "plot", "script"],
    "music": ["music", "song", "sound"],
    "visuals": ["visual", "cinema", "camera"]
}

def extract_theme(text):
    for theme, words in themes.items():
        if any(w in text for w in words):
            return theme
    return "other"

data["keyword_theme"] = data["clean_text"].apply(extract_theme)
data.head()
```

```python
data["length_bucket"] = pd.cut(
    data["word_count"],
    bins=[0,50,150,300,1000],
    labels=["short","medium","long","very long"]
)
print("Length bucket added successfully!")
data.head()
```

```python
df = data
df.to_csv('IMDB_Dataset_Cleaned.csv', index=False)
print("Cleaned dataset saved to 'IMDB_Dataset_Cleaned.csv' successfully!")
```
